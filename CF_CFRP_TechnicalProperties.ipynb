```python
# Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns

plt.rcParams['figure.figsize'] = [100, 100]
plt.rcParams['figure.dpi'] = 300

# Import datasets shown on datatables
df_TorayPanTot = pd.read_csv('FibersTecnProp_ToraycaPAN_Tot.csv', sep=';')
```

```python
df_TorayPanTot
```

```python
x1_product = df_TorayPanTot['Product']
y1_TensileStrength = df_TorayPanTot['Tensile_Strength']
y1_TensileModulus = df_TorayPanTot['Tensile_modulus']
y1_Density = df_TorayPanTot['Density']
y1_Conductivity = df_TorayPanTot['Thermal_Conductivity']
y1_Resistivity = df_TorayPanTot['Electric_Resistivity']
y1_CarbonCont = df_TorayPanTot['Carbon_content']
```

```python
#Create figure element for the drawing space
fig1=plt.figure()
fig1.set_size_inches(20,10)#Set figure size


FiberTypeColor=['lightgrey','lightgrey','lightgrey','darkgrey','darkgrey','darkgrey','darkgrey','dimgray','dimgray','dimgray','dimgray','dimgray']

legend_dict={'Standard Modulus':'lightgrey','Intermediate Modulus':'darkgrey','Ultra High Modulus':'dimgray'}

patchList=[]
for key in legend_dict:
        label_key=mpatches.Patch(color=legend_dict[key],label=key)
        patchList.append(label_key)
        
standard_patch = mpatches.Patch(color='lightgrey', label='Standard Modulus')
intermediate_patch = mpatches.Patch(color='darkgrey', label='Intermediate Modulus')
#high_patch = mpatches.Patch(color='dimgray', label='Ultra High Modulus')
                                
ax1=fig1.add_subplot(3,2,1)#Create de axes artist object (axes container)
ax1.bar(x1_product,y1_TensileStrength,color=FiberTypeColor)#Create bar plot
ax1.set_ylabel('Tensile Strength (MPa)')
ax1.legend(handles=patchList,loc=4)

ax2=fig1.add_subplot(3,2,2)
ax2.bar(x1_product,y1_TensileModulus,color=FiberTypeColor)
ax2.set_ylabel('Tensile Modulus (GPa)')
ax2.legend(handles=patchList,loc=4)

ax3=fig1.add_subplot(3,2,3)
ax3.bar(x1_product,y1_Density,color=FiberTypeColor)
ax3.set_ylabel('Density (g/cm3)')
ax3.legend(handles=patchList,loc=4)

ax4=fig1.add_subplot(3,2,4)
ax4.bar(x1_product,y1_Conductivity,color=FiberTypeColor)
ax4.set_ylabel('Thermal Conductivity (W/mK)')
ax4.legend(handles=patchList,loc=4)

ax5=fig1.add_subplot(3,2,5)
ax5.bar(x1_product,y1_Resistivity,color=FiberTypeColor)
ax5.set_ylabel('Electrical Resistivity (ohmmetere-6)')
ax5.legend(handles=patchList,loc=4)

ax6=fig1.add_subplot(3,2,6)
ax6.bar(x1_product,y1_CarbonCont,color=FiberTypeColor)
ax6.set_ylabel('Carbon Content (%)')
ax6.legend(handles=patchList,loc=4)

plt.tight_layout(pad=1, w_pad=5, h_pad=1)#Adjusts the spacing between plots
```
