```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns

sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = [100, 100]
plt.rcParams['figure.dpi'] = 300

# Import datasets shown on datatables
df_TorayPanTot = pd.read_csv('FibersTecnProp_ToraycaPAN_Tot_multi.csv', sep=';')
```

```python
df_TorayPanTot
```

```python
sns.pairplot(df_TorayPanTot,kind='reg',plot_kws={'line_kws':{'color':'red'}})
```

```python
sns.pairplot(df_TorayPanTot,hue='Type', corner=True)
```

Carbon VS Tensile_strength: Se ve relación positiva hasta standard, después tensile strength se reduce para ultrahigh modulus, puede deberse a una rotura frágil.

Carbon_content VS tensile_modulus: Aquí la relación es directa, a medida que aumenta la candidad de carbono aumenta el módulo elástico.

Carbon_content VS Density: Relación directa, aumenta la densidad con el contenido de carbono (es posible que aumente el packing de los cristales)

Carbon_content VS thermal_conductivity: Relación directa, aumenta el carbono, aumenta la conductividad térmica

Carbon_content VS electric_resistivity: Rlación inversa, aumenta el carbono y disminuye la resistividad eléctrica
